# Импорт библиотек и классов

```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np

sns.set()
```

```python
from can import CAN
```

# Сбор статистики

```python
def sum_stats(stats):
    stats['n_neighbors'] = np.array(stats['n_neighbors'])
    stats['path_length'] = np.array(stats['path_length'])
    stats['n_actions'] = np.array(stats['n_actions'])
    return {'n_neighbors': (stats['n_neighbors'].min(), stats['n_neighbors'].mean(), stats['n_neighbors'].max()),
           'path_length': (stats['path_length'].min(), stats['path_length'].mean(), stats['path_length'].max()),
           'n_actions': (stats['n_actions'].min(axis=1), stats['n_actions'].mean(axis=1), stats['n_actions'].max(axis=1))}
```

# Пробный запуск

Генерация графа:

```python
can = CAN(1000, 42, mode='run', random_pointers=None)
can.generate()
graph_stats = can.visualize(size=512, max_iter=1000, show_names=False, show_path=True, show_frames=True, collect_stats=True)
```

Распределение числа соседей:

```python
plt.figure(figsize=(10,7))
plt.hist(graph_stats['n_neighbors'], bins=6, align='left')
plt.show()
```

10 вершин в графе и 1 запрос:

```python
can = CAN(10, 42, mode='run', random_pointers=None)
can.generate()
can.generate_action(position=(0.55, 0.55), sender_id=0)
can.visualize(size=512, max_iter=1000, show_names=True, show_path=True, show_frames=True)
```

## 1 запрос

```python
can = CAN(1000, 42, mode='run', random_pointers=None)
can.generate()
can.generate_action(position=(0.55, 0.55), sender_id=0)
stats = can.visualize(size=512, max_iter=1000, show_names=False, show_path=True, show_frames=True,
                      collect_stats=True, save_folder="test_1000_1/")


sum_stats(stats)
```

## 100 запросов

```python
can = CAN(1000, 42, mode='run', random_pointers=None)
can.generate()
for i in range(100):
    can.generate_action()
stats = can.visualize(size=512, max_iter=1000, show_names=False, show_path=False, show_frames=False,
                      collect_stats=True, save_folder="test_1000_100/")

sum_stats(stats)
```

Как быстро 100 отправленных в один момент запросов достигнут цели:

```python
plt.figure(figsize=(15,7))
plt.subplot(1, 2, 1)
plt.title('mean number of actions')
plt.plot(sum_stats(stats)['n_actions'][1])
plt.subplot(1, 2, 2)
plt.plot(sum_stats(stats)['n_actions'][2])
plt.title('max number of actions')
plt.show()
```

Распределение длин путей:

```python
plt.figure(figsize=(10,7))
plt.hist(stats['path_length'])
plt.title("Path length")
plt.show()
```
